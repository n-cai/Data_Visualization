```python
movies_clean = movies.dropna(subset=['Distributor','US_Gross'])
print (len(movies_clean))
```

```python
dis_list = movies_clean.Distributor.value_counts().index.tolist()
```

```python
#print(dis_list)

fig, axes = plt.subplots(2, 2, figsize=(15,5),sharex=True, sharey=True)
for axes, dis in zip(axes.flat, dis_list[:4]):
    sns.scatterplot(x='Production_Budget', y='US_Gross', data = movies_clean[movies_clean.Distributor==dis])
plt.show()
```
