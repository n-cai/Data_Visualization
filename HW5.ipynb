```python
import altair as alt
from vega_datasets import data
```

```python
data.list_datasets()
```

```python
iris = data.iris()
iris
```

```python
brush = alt.selection_interval()
points = alt.Chart(iris).mark_point().encode(
    x='sepalLength:Q',
    y='petalLength:Q',
    color='species:N',
    tooltip=['sepalLength','petalLength']
).add_params(brush)

points
```

```python
bar = alt.Chart(iris).mark_bar().encode(
    y='species:N',
    x='count(species):Q',
    color='species:N'
).transform_filter(brush)
```

```python
points & bar
```

```python
brush = alt.selection_interval()
points1 = alt.Chart(iris).mark_point().encode(
    x='SepalRatio:Q',
    y='PetalRatio:Q',
    color='species:N',
).transform_calculate(
    SepalRatio=alt.datum.sepalLength/alt.datum.sepalWidth,
    PetalRatio=alt.datum.petalLength/alt.datum.petalWidth
    ).add_params(brush)

points1
```

```python
hist = alt.Chart(iris).mark_bar().encode(
    x = alt.X('sepalLength:Q', bin=True),
    y='count()'
)
hist
```
